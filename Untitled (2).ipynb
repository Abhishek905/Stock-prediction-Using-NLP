```python
import csv
from nltk.corpus import stopwords
import string
import nltk
from wordcloud import WordCloud, STOPWORDS
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus1 = []
import re
for i in range(0,len(data1.index)):
    rev = re.sub('[^A-Za-z]',' ',data1.iloc[i,0])
    rev = rev.lower().split()
    rev = [ps.stem(word) for word in rev if not word in set(stopwords.words('english'))]
    rev = ' '.join(rev)
    corpus1.append(rev)
```

```python
corpus1[0]
```

```python
freq = pd.Series(' '.join(corpus1).split()).value_counts()
freq
```

```python
freq=pd.DataFrame(freq)
freq =freq.reset_index()
freq.columns=['words','count']
```

```python
#word cloud
wordcloud=WordCloud(width=1000,height=500, stopwords=STOPWORDS, background_color='white').generate(' '.join(freq['words']))
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
corpus2 = pd.Series(corpus1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer(ngram_range = (1,1))
converted = bow.fit_transform(corpus2).toarray()
```

```python
type(converted)
```

```python
train_con = pd.DataFrame(converted)
```

```python
train_con.columns = bow.get_feature_names()
```

```python
train_con.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_classifier = RandomForestClassifier(n_estimators= 200, criterion = 'entropy')
RF_classifier.fit(train_con,train['Label'])
```

```python
data2= test.iloc[:,2:27]
```

```python
test_cor = []
for i in range(0,len(data2.index)):
  x=' '.join(str(word) for word in data2.iloc[i,0:25])
  test_cor.append(x)
```

```python
testdata = bow.transform(test_cor)
```

```python
predict = RF_classifier.predict(testdata)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix= confusion_matrix(test['Label'],predict)
matrix
```

```python
test['Label'].value_counts()
```

```python
score = accuracy_score(test['Label'],predict)
score
```

```python
report = classification_report(test['Label'],predict)
report
```
